## Data preprocessing

```python
plt.subplots(figsize = (10,5))
sns.countplot("Class",data=data_df)
```

```python
# now let us check in the number of Percentage
Count_Normal_transacation = len(data_df[data_df["Class"]==0]) # normal transaction are repersented by 0
Count_Fraud_transacation = len(data_df[data_df["Class"]==1]) # fraud by 1
Percentage_of_Normal_transacation = Count_Normal_transacation/(Count_Normal_transacation+Count_Fraud_transacation)
print("percentage of normal transacation is", '%s%%'%(Percentage_of_Normal_transacation*100))
Percentage_of_Fraud_transacation= Count_Fraud_transacation/(Count_Normal_transacation+Count_Fraud_transacation)
print("percentage of fraud transacation",'%s%%'%(Percentage_of_Fraud_transacation))
```

highly imbalanced dataset

```python
# check missing value
plt.subplots(figsize = (10,5))
sns.heatmap(data_df.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

Good! No missing value

```python
class_0 = data_df.loc[data_df['Class'] == 0]["Time"]
class_1 = data_df.loc[data_df['Class'] == 1]["Time"]
plt.subplots(figsize = (10,5))
plt.title('Credit Card Transactions Time Density Plot')
sns.distplot(class_0,hist = False, kde = True, kde_kws = {'shade': True, 'linewidth': 3}, label = 'Not Fraud')
sns.distplot(class_1,hist = False, kde = True, kde_kws = {'shade': True, 'linewidth': 3}, label = 'Fraud')
```

```python
Fraud = data_df.loc[data_df['Class'] == 1]
plt.subplots(figsize = (10,5))
sns.scatterplot(x=Fraud["Time"], y=Fraud["Amount"])
```

```python
correlations = data_df.corr()
plt.subplots(figsize = (14, 14))
plt.title('Credit Card Transactions features correlation plot')
sns.heatmap(correlations, cmap="YlGnBu")
```
